```python
import numpy as np
import pandas as pd
```

#**Aritmethic Operation**

```python
a = np.arange(9, dtype = np.float_).reshape(3,3)

print('Array pertama:')
print(a)
print('\n')

print('Array kedua:')
b = np.array([10,10,10])
print(b)
print('\n')

print('Penjumlahan 2 array:')
print(np.add(a,b))
print('\n')

print('Pengurangan 2 array:')
print(np.subtract(a,b))
print('\n')

print('Perkalian 2 array:')
print(np.multiply(a,b))
print('\n')

print('Pembagian 2 array:')
print(np.divide(a,b))
```

#**Exponential Operation**

```python
a = np.array([110, 100, 90, 80])
b = np.array([1,    2,   3,  4])

print("original array : ")
print(a)

print("using power function : ")
print(np.power(a,2))

print("\nsecond array : ")
print(b)

print("using power function : ")
print(np.power(a, b))
```

#**Modulus Operation**

```python
a = np.array([10, 20, 30, 40])
b = np.array([3, 5, 7, 9])

print("first array : ")
print(a)

print("second array : ")
print(b)


print("\nmodulus function : ")
print(np.mod(a,b))

print("remainder function : ")
print(np.remainder(a, b))
```

#**Statistical Function**

```python
a = np.array([[3,7,5],[2,4,3],[9,8,7]])

print("original array : ")
print(a)

print("\nminimun value from array : ")
print(np.min(a))


print("\nmin value based on row : ")
print(np.amin(a, 1)) #axis = 1 -> berdasarkan baris
                     #axis = 0 -> berdasarkan column
                     #so the output read as berdasarkan row, nilai mana yang paling kecil

print("\nminimum value based on column : ")
print(np.amin(a, 0))

```

```python

print("\nmaximum value from array : ")
print(np.max(a))

print("\nmaximum value based on row : ")
print(np.amax(a, 1))

print("\nmaximum value based on column : ")
print(np.amax(a, 0))
```

##**Percentile**

```python
a = np.array([[90, 80 ,70 ], [89, 70, 90], [85, 75, 77]])

print("original array : ")
print(a)

print("\n50% percentile : ")
print(np.percentile(a, 50))

print("\n50% percentile di sumbu 1  : ")
print(np.percentile(a, 50, axis=1))

print("\n50% percentile di sumbu 0  : ")
print(np.percentile(a, 50, axis=0))
```

##**Weighted Average**

Rata-rata tertimbang adalah rata-rata yang dihasilkan dari perkalian setiap komponen dengan faktor yang mencerminkan kepentingannya. Fungsi numpy.average() menghitung rata-rata tertimbang elemen dalam array sesuai dengan bobot masing-masing yang diberikan dalam array lain. Fungsi dapat memiliki parameter sumbu. Jika sumbu tidak ditentukan, array diratakan.

Mempertimbangkan array [1,2,3,4] dan bobot yang sesuai [4,3,2,1], rata-rata tertimbang dihitung dengan menambahkan produk dari elemen yang sesuai dan membagi jumlah dengan jumlah bobot.

Rata-rata tertimbang = (1x4 + 2x3 + 3x2 + 4x1)/(4+3+2+1)

```python
a = np.array([[90, 80 ,70, 60, 65 ]])

print("original array : ")
print(a)

print("average function : ")
print(np.average(a))

#lalu, tentukan weight dr array

weight = np.array([[3, 5, 7, 1, 2]])

#read it as = 90 ada 3
#             80 ada 5
#             dst

#using weighted average function :
print("\nweighted average function : ")
print(np.average(a, weights = weight))
```
